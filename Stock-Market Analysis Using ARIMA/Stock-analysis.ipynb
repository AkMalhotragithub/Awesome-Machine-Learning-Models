```python
#introducing relevant libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
from pandas import datetime
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
#reading the relevant dataset for stocks of Reliance limited ,taken from Yahoo finances
df = pd.read_csv("RELIANCE.csv")
df.head(5)
```

```python
#checking if autocorrection is relevant for the stock price of Reliance limited and whether ARIMA is applicable
plt.figure()
lag_plot(df['Open'], lag=3)
plt.title('RELIANCE LTD Stock- AUTOCORRECTION PLOT(lag=3)')
plt.show()
```

```python
#plotting graph trends for the stock based on previous data
plt.plot(df["Date"], df["Close"])
plt.xticks(np.arange(0,248,50), df['Date'][0:248:50])
plt.title("RELIANCE LTD Stock price vs time")
plt.xlabel("Date")
plt.ylabel("price")
plt.show()
```

```python
#now training model
#dividing data in training set and test set(70 and 60 percent respectively)
train_data, test_data = df[0:int(len(df)*0.7)], df[int(len(df)*0.7):]
training_data = train_data['Close'].values
test_data = test_data['Close'].values
```

```python
history = [x for x in training_data]
model_predictions = []
N_test_observations = len(test_data)
```

```python
#ARIMA code with parameters: p=4, d=1 and q=0.
for time_point in range(N_test_observations):
    model = ARIMA(history, order=(4,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    model_predictions.append(yhat)
    true_test_value = test_data[time_point]
    history.append(true_test_value)
```

```python
#testing mean squared error calculation
MSE_error = mean_squared_error(test_data, model_predictions)
print('Testing Mean Squared Error for this model is {}'.format(MSE_error))
```

```python
#Visualising results and final prediction
test_set_range = df[int(len(df)*0.7):].index
plt.plot(test_set_range, model_predictions, color='blue', marker='o', linestyle='dashed',label='Predicted Price')
plt.plot(test_set_range, test_data, color='green', label='Actual Price')
plt.title('Reliance Ltd. Prices Prediction')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.xticks(np.arange(180,250,15), df.Date[180:250:15])
plt.legend()
plt.show()
```
